```python

ax = sns.countplot(x='Gender',data= df)

for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_gender =df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=True)
# sales_gender
# sales_gender = df.groupby("Gender", as_index=False)["Amount"].sum()
# sales_gender = sales_gender.loc[sales_gender.sort_values(by="Amount",ascending=True).index]

print(sales_gender)

sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(data=sales_gender,x='Gender',y="Amount")
```

```python
gp1 =df.groupby("Gender").agg({"Amount":"sum"}).sort_values(by="Amount",ascending=False)
gp1
# sns.set(rc={"figure.figsize":(10,6)})
# sns.barplot(data=gp1,x="Gender",y="Amount")
# gp1_filtered = gp1.loc[:, ["Amount"]] 
gp1_filtered = gp1.reset_index()[["Gender", "Amount"]]
print(gp1_filtered)
```

```python
s_amount=df.groupby(['State'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,6)})
sns.barplot(x="State",y="Amount",data=s_amount)
```
